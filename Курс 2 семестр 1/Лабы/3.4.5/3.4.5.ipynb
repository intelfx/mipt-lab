```python
import pandas as pd
import math as m
from IPython.display import display as disp
import lablib as ll
%matplotlib inline
```

```python
data, columns, experiments = ll.read_standard_layout()
```

```python
for e in experiments:
    ll.compute(name = "I_c",
               expr = lambda _2x_c, K_x, R_0: (K_x * _2x_c) / 2 / R_0,
               columns = columns[e],
               data = data[e])

    ll.compute(name = "U_s",
               expr = lambda _2y_s, K_y: (K_y * _2y_s) / 2,
               columns = columns[e],
               data = data[e])
    
    ll.compute(name = "H_c",
               expr = lambda I_c, N_0, _2piR: (I_c * N_0) / _2piR,
               columns = columns[e],
               data = data[e])

    ll.compute(name = "B_s",
               expr = lambda U_s, R_rc, C_rc, N_rc, S: (R_rc * C_rc) / (S * N_rc) * U_s,
               columns = columns[e],
               data = data[e])

```

```python
ll.plotfit(title = "B(H)",
           xlabel = "$H_c$, $\\frac{A}{m}$", ylabel = "$B_{sat}, Tl$", axis = [0, None, 0, None],
           x = "H_c", y = "B_s",
           columns = [columns[e] for e in experiments], data = [data[e] for e in experiments], label = experiments,
           plot1_args = { "linestyle": "-" })
           
           
```
