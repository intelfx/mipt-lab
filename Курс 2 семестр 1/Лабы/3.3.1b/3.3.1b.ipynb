```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib as mp
import matplotlib.pyplot as plt
import math as m
import sympy as smp
from IPython.display import display as disp
import lablib as ll
%matplotlib inline
mp.rc("savefig", dpi=96)
```

Загружаем данные. Также заметим, что деление шкалы амперметра в цепи питания соленоида — 0.03 A.

```python
data, columns, experiments = ll.read_standard_layout()

I_f_unit = 0.03 # A per unit

for e in experiments:
    print("%s:" % e)
    disp(data[e], columns[e])
```

Построим графики зависимости анодного тока $I_a$ от тока соленоида $I_f$ при каждом напряжении на аноде ($V_a$).
Визуально определим в каждом случае критический ток через соленоид $I_{cr}$, при котором наблюдается резкое падение анодного тока.

```python
plt.title("Anode current $I_a$ by solenoid current $I_f$")
plt.xlabel("$I_f$")
plt.ylabel("$I_a$")

for e in experiments:
    plt.errorbar(x = columns[e]["I_f"],
                 xerr = columns[e]["Error_I_f"],
                 y = columns[e]["I_a"],
                 yerr = columns[e]["Error_I_a"],
                 label = e)
    
plt.legend(loc = "best")
plt.axis([0, None, 0, None])
None
```


Зафиксируем значения $I_{cr}$ для каждого напряжения на аноде.

```python
ll.add(data["V_a=60"], ll.var(name = "I_cr", value = 40, error = 5))
ll.add(data["V_a=80"], ll.var(name = "I_cr", value = 45, error = 5))
ll.add(data["V_a=100"], ll.var(name = "I_cr", value = 52.5, error = 5))
ll.add(data["V_a=120"], ll.var(name = "I_cr", value = 60, error = 5))
```

Теперь будем строить зависимость критического тока $I_{cr}$ от напряжения на аноде $V_a$. Сначала рассчитаем соответствующие критическим токам напряжённости магнитного поля в соленоиде: $B_{cr} = I_{cr} * \frac{dB}{dI}$ (значение коэффициента известно заранее).

```python
e = "global"

columns[e] = pd.DataFrame()

ll.make_column(where = columns[e], name = "V_a", varlists = data, indices = experiments)
ll.make_column(where = columns[e], name = "I_cr", varlists = data, indices = experiments)

ll.compute(name = "B_cr",
           expr = lambda I_cr, _dB_dI: I_cr * I_f_unit * _dB_dI,
           data = data[e],
           columns = columns[e])

disp(columns[e])
```

Теперь построим график зависимости $B_{cr}^2 = f(V_a)$ и приблизим его прямой, получая коэффициент $\frac{d(B_cr{2})}{d(V_a)}$.

```python
ll.compute(name = "B_cr_sq",
           expr = lambda B_cr: B_cr ** 2,
           data = data[e],
           columns = columns[e])

model = ll.fit2(name = "B_cr_sq",
                model = lambda V_a, _dB_cr_sq_dV_a, _B_cr_sq_0: V_a * _dB_cr_sq_dV_a + _B_cr_sq_0,
                x = columns[e]["V_a"],
                xerr = columns[e]["Error_V_a"],
                y = columns[e]["B_cr"],
                yerr = columns[e]["Error_B_cr"],
                data = data[e])

linspace = ll.linspace(columns[e]["V_a"])

plt.title("Critical field strength $B_{cr}$ by anode voltage $V_a$")
plt.xlabel("$V_a, V$")
plt.ylabel("$B_{cr}, Tl$")

plt.errorbar(x = columns[e]["V_a"],
             xerr = columns[e]["Error_V_a"],
             y = columns[e]["B_cr"],
             yerr = columns[e]["Error_B_cr"],
             linestyle = "none")
plt.errorbar(x = linspace, y = model(linspace))
None
```

Наконец, рассчитаем удельный заряд электрона: $\frac{e}{m} = \frac{8 * V_a}{B_{cr}^2 * R_a^2}$.

```python
ll.compute(name = "_e_m",
           expr = lambda R_a, _dB_cr_sq_dV_a: 8 / R_a**2 / _dB_cr_sq_dV_a,
           data = data[e],
           columns = columns[e])

disp(data[e].loc[["_e_m"]])
```
